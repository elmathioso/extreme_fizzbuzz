```python
for i in range(1, 101):
    word = ''
    if i % 3 == 0: word += 'Fizz'
    if i % 5 == 0: word += 'Buzz'
    print(word) if len(word) > 0 else print(i)
        
```

```python
for i in range(1, 101):

    # filter for both fizzbuzz first
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")

    # then evaluate either option
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")

    # then print the number
    else: 
        print(i)
```

```python
fizzlist = ['Fizz' if x % 3 == 0 else '' for x in range(1, 101)]
buzzlist = ['Buzz' if x % 5 == 0 else '' for x in range(1, 101)]
fizzbuzz = {k+1: fizzlist[k] + buzzlist[k] for k in range(len(fizzlist))}

for i in fizzbuzz.items():
    print(i[1] if len(i[1]) > 0 else i[0])
```
